```python
# 按照1分钟的时间间隔，对事件的序列进行分割，并对分割后的每一个子序列进行检测


def get_input_events_by_minute(whole_df):
    
    time_span = dt.timedelta(seconds=60)
    
    input_df_list = []
    
    nrow, ncol = whole_df.shape
    
    start_time = whole_df.iloc[0].datetime
    temp_time = start_time
    k = 0
    
    
    for i in tqdm(range(nrow)):
        
        temp_time = whole_df.iloc[i].datetime
        
        if (start_time + time_span ) <= temp_time:
            
            df_input = whole_df[k:i]
            input_df_list.append(df_input)
            k = i
            start_time = temp_time
            
    df_input = whole_df[k:]
    input_df_list.append(df_input)
    
    
    return input_df_list
```

```python
def randomtimes(stime, etime):
     
    return random.random() * (etime - stime) + stime 
 
```

```python
# 这是生成随机异常序列的实现函数

# input_df_list : 输入的以一分钟为时间间隔的序列
# events_total : 事件的总数量（所有序列）
# percentage ： 异常事件的百分比
# sensors_number : 传感器总数
# all_sensor_id_list 设备名称列表

events_total, ncol = test_seg_df.shape

def attack_random(input_df_list, events_total, percentage, all_sensor_id_list ):
    
 
    input_len = len(input_df_list)
    sensors_number = len(all_sensor_id_list)
     
    attacked_df = pd.DataFrame(columns = ['datetime', 'sensor_id', 'val', 'flag'])
    
    # start_time =     target_df.iloc[0].datetime
    # temp_time = start_time
    
    for i in tqdm(range( int(percentage * events_total))):
        
        seg_id = np.random.randint(0, input_len)
        
        temp_df = input_df_list[seg_id]
        temp_sensors_list = temp_df.sensor_id.unique().tolist()
        
        temp_start_time = input_df_list[seg_id].iloc[0].datetime
        temp_end_time = input_df_list[seg_id].iloc[-1].datetime
        
        temp_time_stamp = randomtimes(input_df_list[seg_id].iloc[0].datetime, input_df_list[seg_id].iloc[-1].datetime)
        
        found = False
        while not found:
            
            sensor_id = np.random.randint(0, sensors_number)
            sensor_name = all_sensor_id_list[sensor_id]
            
            if sensor_name not in set(temp_sensors_list) and (sensor_name.startswith('M') or sensor_name.startswith('D')):
            
                found = True
                
         
        one_a_df = temp_df.iloc[0:1].copy()
        one_a_df.sensor_id = sensor_name
        one_a_df.val = np.random.randint(0, 2)
        one_a_df.datetime = temp_time_stamp
        one_a_df.flag = 1

        attacked_df = pd.concat([attacked_df, one_a_df], axis=0)        
 
                    
    return attacked_df
```

```python
attacked_df = attack_random(input_df_list, events_total, 0.1, all_sensor_id_list)
```
